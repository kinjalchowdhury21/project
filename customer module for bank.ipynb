```python
name_list=[]
acc_no_list=[]
phoneno_list=[]
email_list=[]
customer_info={}
while True:
    name=input('enter your name:')
    acc_no=input('enter your acc no:')
    phone_no=input('enter the phone no:')
    email_id=input('enter your email:')
    
    
    def check_acc_len(acc_no):       
        if 7<=len(acc_no)<=12:
            return True
        return False

    def check_acc_first3char(acc_no):
        for i in range(0,3):
            if not('A'<= acc_no[i] <= 'Z' or 'a'<=acc_no[i]<='z'):
                return False
        return True
    def check_start_alpha(acc_no):
        if acc_no[0] in 'scfrdSCFRD':
            return True
        return False
    def check_no_in_acc(acc_no):
        for i in range(3,(len(acc_no)-1)):
            if not('0'<=acc_no[i]<='9'):
                return False
        return True
    
        
    def check_cust_name(name):
        for i in name:
            if not('A'<=i<='Z' or 'a'<=i<='z'):
                return False
        return True
        
        
     


    def check_phoneno(phone_no):
        if len(phone_no)==10:
            for i in phone_no:
                if not('0'<=i<='9'):
                    return False
            return True
    def check_email(email_id):
        if email_id[0]=='@' or  email_id[-1]=='@':
            return False
        count=0
        for i in range(1,len(email_id)):
            if email_id[i]=='@':
                count+=1
        if count==1:
            return True
        else:
            False
            
            


     
    if check_phoneno(phone_no)==True:
        phoneno_list+=[phone_no]
    else:
        print('invalid phoneno')
        break

    if check_email(email_id)==True:
        email_list+=[email_id]
    else:
        print('invalid emailid')
        break

    if check_cust_name(name)==True:
        name_list+=[name]
    else:
        print('invalid name')
        break

    if check_no_in_acc(acc_no)==True and check_start_alpha(acc_no)==True and check_acc_first3char(acc_no)==True and check_acc_len(acc_no)==True:
        acc_no_list+=[acc_no]
    else:
        print('invalid acc no')
        break



    print(phoneno_list)
    print(email_list)
    print(name_list)
    print(acc_no_list)

    for i in acc_no_list:
        if 'acc_list' in customer_info:
            customer_info['acc_list']+=[i]
        else:
            customer_info['acc_list']=[i]
    for j in name_list:
        if 'customer_name_list' in customer_info:
            customer_info['customer_name_list']+=[j]
        else:
            customer_info['customer_name_list']=[j]
    for k in phoneno_list:
        if 'phoneno_list' in customer_info:
            customer_info['phoneno_list']+=[k]
        else:
            customer_info['phoneno_list']=[k]
    for l in email_list:
        if 'emailid_list' in customer_info:
            customer_info['emailid_list']+=[l]
        else:
            customer_info['emailid_list']=[l]


print('your customer info dictionary is',customer_info)

import pandas as pd
cust=pd.DataFrame(customer_info)
cust.to_csv("kincustomer_info.csv")



    
```
